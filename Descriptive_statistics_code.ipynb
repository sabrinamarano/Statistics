```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew

# Simulated daily ad clicks (slightly skewed and leptokurtic)
np.random.seed(42)
ad_clicks = np.random.normal(1000, 100, 1000) + np.random.exponential(50, 1000)

# Calculate skewness and kurtosis
data_skewness = skew(ad_clicks)
data_kurtosis = kurtosis(ad_clicks, fisher=False)  # Fisher=False gives Pearson kurtosis (adds 3)

# Plot the distribution
sns.histplot(ad_clicks, kde=True, bins=30, color="skyblue")
plt.title(f"Ad Clicks Distribution\nSkewness: {data_skewness:.2f}, Kurtosis: {data_kurtosis:.2f}")
plt.xlabel("Daily Ad Clicks")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Marketing data: Customer spending in dollars
mean = 50  # Mean spending
std_dev = 15  # Standard deviation of spending

# Generate data for 1000 customers
data = np.random.normal(loc=mean, scale=std_dev, size=1000)

# Create a figure and axis for plotting
plt.figure(figsize=(10,6))

# Plot the histogram of the data
sns.histplot(data, kde=True, stat="density", color="blue", bins=30, label="Customer Spending")

# Add the normal distribution curve
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
plt.plot(x, norm.pdf(x, mean, std_dev), color='red', label="Normal Distribution", lw=2)

# Calculate the probability of spending less than $35
prob_less_35 = norm.cdf(35, mean, std_dev)
prob_more_65 = 1 - norm.cdf(65, mean, std_dev)

# Highlight the area where spending is less than $35 (left side)
x_fill_1 = np.linspace(mean - 4*std_dev, 35, 500)
plt.fill_between(x_fill_1, norm.pdf(x_fill_1, mean, std_dev), color='blue', alpha=0.3)

# Highlight the area where spending is more than $65 (right side)
x_fill_2 = np.linspace(65, mean + 4*std_dev, 500)
plt.fill_between(x_fill_2, norm.pdf(x_fill_2, mean, std_dev), color='red', alpha=0.3)

# Add labels and title
plt.title("Normal Distribution of Customer Spending", fontsize=16)
plt.xlabel("Spending in Dollars", fontsize=12)
plt.ylabel("Density", fontsize=12)

# Annotate the percentage areas
plt.text(20, 0.02, f"Less than $35\n{prob_less_35*100:.2f}%", fontsize=12, color="blue")
plt.text(75, 0.02, f"More than $65\n{prob_more_65*100:.2f}%", fontsize=12, color="red")

# Add legend
plt.legend()

# Show the plot
plt.show()

# Print the probability results
print(f"Percentage of customers spending less than $35: {prob_less_35 * 100:.2f}%")
print(f"Percentage of customers spending more than $65: {prob_more_65 * 100:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the normal distribution
mean = 0
std_dev = 1

# Generate data for plotting the normal distribution curve
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = norm.pdf(x, mean, std_dev)

# Create the plot
plt.figure(figsize=(10,6))
plt.plot(x, y, label="Normal Distribution", color='b')

# Fill areas corresponding to the Empirical Rule
# 1 Standard Deviation Range (68%)
plt.fill_between(x, y, where=(x >= mean - std_dev) & (x <= mean + std_dev), color='skyblue', alpha=0.4, label='±1 SD (68%)')

# 2 Standard Deviations Range (95%)
plt.fill_between(x, y, where=(x >= mean - 2*std_dev) & (x <= mean + 2*std_dev), color='orange', alpha=0.4, label='±2 SD (95%)')

# 3 Standard Deviations Range (99.7%)
plt.fill_between(x, y, where=(x >= mean - 3*std_dev) & (x <= mean + 3*std_dev), color='green', alpha=0.4, label='±3 SD (99.7%)')

# Add vertical lines for the mean and standard deviations
plt.axvline(mean, color='black', linestyle='dashed', label='Mean (0%)')
plt.axvline(mean - std_dev, color='red', linestyle='dotted', label='-1 SD')
plt.axvline(mean + std_dev, color='red', linestyle='dotted', label='+1 SD')
plt.axvline(mean - 2*std_dev, color='purple', linestyle='dotted', label='-2 SD')
plt.axvline(mean + 2*std_dev, color='purple', linestyle='dotted', label='+2 SD')
plt.axvline(mean - 3*std_dev, color='brown', linestyle='dotted', label='-3 SD')
plt.axvline(mean + 3*std_dev, color='brown', linestyle='dotted', label='+3 SD')

# Add labels and title
plt.title("Normal Distribution with Empirical Rule", fontsize=14)
plt.xlabel("Data Value", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Generating data
np.random.seed(42)

# Mesokurtic (Normal Distribution)
normal_data = np.random.normal(loc=50, scale=10, size=1000)

# Leptokurtic (High Kurtosis)
leptokurtic_data = np.random.laplace(loc=50, scale=10, size=1000)

# Platykurtic (Low Kurtosis)
platykurtic_data = np.random.uniform(low=30, high=70, size=1000)

# Calculate Kurtosis
normal_kurt = kurtosis(normal_data, fisher=False)
lepto_kurt = kurtosis(leptokurtic_data, fisher=False)
platy_kurt = kurtosis(platykurtic_data, fisher=False)

# Plotting
data_sets = [normal_data, leptokurtic_data, platykurtic_data]
titles = [
    f"Mesokurtic (Kurtosis: {normal_kurt:.2f})",
    f"Leptokurtic (Kurtosis: {lepto_kurt:.2f})",
    f"Platykurtic (Kurtosis: {platy_kurt:.2f})"
]

plt.figure(figsize=(15, 5))
for i, data in enumerate(data_sets):
    plt.subplot(1, 3, i + 1)
    plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title(titles[i])
    plt.xlabel('Value')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data for each distribution

# 1. Normal Distribution
normal_data = np.random.normal(loc=50, scale=15, size=1000)  # mean=50, std=15

# 2. Right Skewed Distribution
right_skewed_data = np.random.beta(a=2, b=5, size=1000) * 50 + 25  # Right skewed

# 3. Left Skewed Distribution
left_skewed_data = np.random.beta(a=5, b=2, size=1000) * 50 + 25  # Left skewed

# Create a figure with multiple subplots
plt.figure(figsize=(12, 8))

# Plot the boxplot for normal distribution
plt.subplot(131)
sns.boxplot(data=normal_data, color='skyblue')
plt.title('Normal Distribution')

# Plot the boxplot for right skewed distribution
plt.subplot(132)
sns.boxplot(data=right_skewed_data, color='lightgreen')
plt.title('Right Skewed Distribution')

# Plot the boxplot for left skewed distribution
plt.subplot(133)
sns.boxplot(data=left_skewed_data, color='lightcoral')
plt.title('Left Skewed Distribution')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data (e.g., for a general dataset)
data = np.random.normal(loc=50, scale=15, size=1000)  # mean=50, std=15

# Create a figure for the boxplot
plt.figure(figsize=(8, 6))

# Create the boxplot using seaborn
sns.boxplot(data=data, color='skyblue')

# Get the summary statistics for the boxplot
Q1 = np.percentile(data, 25)  # 25th percentile (Q1)
Q3 = np.percentile(data, 75)  # 75th percentile (Q3)
median = np.median(data)  # Median (50th percentile)
whisker_lower = np.percentile(data, 5)  # Lower whisker (5th percentile)
whisker_upper = np.percentile(data, 95)  # Upper whisker (95th percentile)

# Annotate the boxplot components
plt.text(0, Q1, f'Q1: {Q1:.2f}', horizontalalignment='center', color='red', fontsize=12)
plt.text(0, Q3, f'Q3: {Q3:.2f}', horizontalalignment='center', color='red', fontsize=12)
plt.text(0, median, f'Median: {median:.2f}', horizontalalignment='center', color='green', fontsize=12)
plt.text(0, whisker_lower, f'Whisker Lower: {whisker_lower:.2f}', horizontalalignment='center', color='purple', fontsize=12)
plt.text(0, whisker_upper, f'Whisker Upper: {whisker_upper:.2f}', horizontalalignment='center', color='purple', fontsize=12)

# Add title and labels
plt.title('Boxplot Highlighting Key Components', fontsize=14)
plt.xlabel('Data', fontsize=12)
plt.ylabel('Values', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate customer response times to an email promotion (in hours)
np.random.seed(42)  # For reproducibility
response_times = np.random.normal(loc=5, scale=2, size=500)  # mean=5 hours, std=2 hours

# Create a figure for the boxplot
plt.figure(figsize=(8, 6))

# Create the boxplot using seaborn
sns.boxplot(data=response_times, color='lightgreen')

# Add title and labels
plt.title('Boxplot of Customer Response Times to Email Promotion', fontsize=14)
plt.xlabel('Response Time (in hours)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Simulate satisfaction scores for different products
np.random.seed(42)
product_a = np.random.normal(loc=75, scale=10, size=100)
product_b = np.random.normal(loc=80, scale=15, size=100)
product_c = np.random.normal(loc=70, scale=5, size=100)
# Create a DataFrame for plotting
import pandas as pd
df = pd.DataFrame({
    "Product A": product_a,
    "Product B": product_b,
    "Product C": product_c
})
# Create box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df)
plt.title("Box Plot of Product Satisfaction Scores")
plt.xlabel("Products")
plt.ylabel("Satisfaction Score")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate data (e.g., customer age and customer spending)
np.random.seed(42)
age = np.random.randint(18, 70, size=100)  # Age between 18 and 70
spending = 50 + age * 0.8 + np.random.normal(0, 10, size=100)  # Spending correlated with age

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(age, spending, color='skyblue', edgecolor='black')
plt.title("Scatter Plot of Customer Age vs. Spending")
plt.xlabel("Age")
plt.ylabel("Spending ($)")
plt.show()

# Calculate the correlation coefficient (r)
correlation = np.corrcoef(age, spending)[0, 1]
print("Correlation coefficient (r):", correlation)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulated data: User activity across the hours of the day (0 to 23) for 7 days of the week
np.random.seed(42)
hours = np.arange(24)  # 0 to 23 hours
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Generate random user activity (values between 0 and 100)
activity_data = np.random.randint(0, 100, size=(7, 24))

# Create a DataFrame for easy manipulation
df_activity = pd.DataFrame(activity_data, columns=hours, index=days)

# Create a heatmap to visualize user activity
plt.figure(figsize=(10, 8))
sns.heatmap(df_activity, annot=True, cmap='YlGnBu', cbar_kws={'label': 'User Activity'}, linewidths=0.5)
plt.title('User Activity by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Day of the Week')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (e.g., customer spending data)
np.random.seed(42)
customer_spending = np.random.normal(loc=50, scale=15, size=1000)

# Create a density plot
plt.figure(figsize=(8, 6))
sns.kdeplot(customer_spending, shade=True, color="blue", linewidth=2)

# Add titles and labels
plt.title('Density Plot of Customer Spending')
plt.xlabel('Spending ($)')
plt.ylabel('Density')

# Show the plot
plt.show()
```
